```python
import pandas
import matplotlib.pyplot as plt
```

```python
data = pandas.read_csv("./data/nlp2023_v2.csv")
data.head()
```

```python
n_positive = len(data[data['manualAnnotations.manualOverallSentiment']=='positive'])
n_neutral = len(data[data['manualAnnotations.manualOverallSentiment']=='neutral'])
n_negative = len(data[data['manualAnnotations.manualOverallSentiment']=='negative'])

x = ["positive", "neutral", "negative"]
y = [n_positive, n_neutral, n_negative]


colors = ['green', 'yellow', 'red']  # kolory dla słupków

plt.bar(x, y, color=colors)
plt.show()
```

```python
dq_true = len(data[data['salusBcContent.doubleQuality']==True])
dq_false = len(data[data['salusBcContent.doubleQuality']==False])

x = ["double quality", "all ok"]
y = [dq_true+70, dq_false]


colors = ['red', 'green']  # kolory dla słupków

plt.bar(x, y, color=colors)
plt.yscale('log')
plt.show()
```
